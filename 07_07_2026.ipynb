```python
list1 = ["j", "d", "w", "f", "n", "22"]
print("Original List:", list1)

print("\n1. append()")
list2 = list1.copy()
list2.append("k")
print(list2)

print("\n2. extend()")
list2 = list1.copy()
list2.extend(["a", "b", "c"])
print(list2)

print("\n3. insert()")
list2 = list1.copy()
list2.insert(2, "x")
print(list2)

print("\n4. remove()")
list2 = list1.copy()
list2.remove("22")
print(list2)

print("\n5. pop()")
list2 = list1.copy()
removed = list2.pop()
print("Removed Element:", removed)
print("Updated List:", list2)

print("\n6. clear()")
list2 = list1.copy()
list2.clear()
print(list2)

print("\n7. del")
list2 = list1.copy()
print("Before del:", list2)
del list2
print("List deleted successfully.")

print("\n8. index()")
print("Index of 'w' =", list1.index("w"))

print("\n9. count()")
print("Count of '22' =", list1.count("22"))

print("\n10. sort()")
list2 = list1.copy()
list2.sort()
print(list2)

print("\n11. reverse()")
list2 = list1.copy()
list2.reverse()
print(list2)
```
